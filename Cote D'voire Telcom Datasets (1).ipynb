```python
# converting our data types to the required form in the description
cells_geo['VILLES']=cells_geo['VILLES'].astype('string')
cells_geo['STATUS']=cells_geo['STATUS'].astype('string')
cells_geo['LOCALISATION']=cells_geo['LOCALISATION'].astype('string')
cells_geo['DECOUPZONE']=cells_geo['DECOUPZONE'].astype('string')
cells_geo['ZONENAME']=cells_geo['ZONENAME'].astype('string')
cells_geo['REGION']=cells_geo['REGION'].astype('string')
cells_geo['AREA']=cells_geo['AREA'].astype('string')
cells_geo['CELL_ID']=cells_geo['CELL_ID'].astype('string')
cells_geo['SITE_CODE']=cells_geo['SITE_CODE'].astype('string')
```

# Data Analysis

```python
New_df['PRODUCT'].value_counts()
```

```python
#which product was was most and less used in what country

New_df[New_df['PRODUCT'] == 'sms']['VILLES'].value_counts()
```
